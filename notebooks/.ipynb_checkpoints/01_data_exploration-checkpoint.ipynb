```python
"""
01_data_exploration.ipynb
Explore the loaded data to understand structure, missing values, patterns
"""

import pandas as pd
import matplotlib.pyplot as plt
import sys
import os

sys.path.append(os.path.dirname(os.path.dirname(os.path.abspath(__file__))))

from pipeline.db import engine

# ─── 1. Check what projects were loaded ──────────────────────────────────────
print("=" * 60)
print("STEP 1: Check loaded projects")
print("=" * 60)

projects = pd.read_sql("SELECT * FROM projects", engine)
print(f"Loaded {len(projects)} projects:")
print(projects[['project_id', 'project_name', 'budget_at_completion', 'total_periods']])

# ─── 2. Check period data for each project ───────────────────────────────────
print("\n" + "=" * 60)
print("STEP 2: Period data overview")
print("=" * 60)

period_counts = pd.read_sql("""
    SELECT p.project_id, p.project_name, COUNT(per.period_id) as period_count
    FROM projects p
    LEFT JOIN periods per ON p.project_id = per.project_id
    GROUP BY p.project_id, p.project_name
    ORDER BY p.project_id
""", engine)

print(period_counts)

# ─── 3. Deep dive into C2011-13 (the 119-period project) ─────────────────────
print("\n" + "=" * 60)
print("STEP 3: Deep dive into C2011-13 Wind Farm (119 periods)")
print("=" * 60)

# Get periods for C2011-13
periods = pd.read_sql("""
    SELECT * FROM periods 
    WHERE project_id = (
        SELECT project_id FROM projects WHERE project_name LIKE '%Wind%'
    )
    ORDER BY period_number
""", engine)

print(f"Periods loaded: {len(periods)}")
print("\nFirst 10 periods:")
print(periods[['period_number', 'planned_value', 'earned_value', 'actual_cost']].head(10))

print("\nLast 10 periods:")
print(periods[['period_number', 'planned_value', 'earned_value', 'actual_cost']].tail(10))

# ─── 4. Check for missing values ─────────────────────────────────────────────
print("\n" + "=" * 60)
print("STEP 4: Missing values check")
print("=" * 60)

missing = periods[['planned_value', 'earned_value', 'actual_cost']].isnull().sum()
print(f"Missing values in C2011-13:")
print(missing)

# ─── 5. Check for negative values ────────────────────────────────────────────
print("\n" + "=" * 60)
print("STEP 5: Negative values check")
print("=" * 60)

negatives = periods[
    (periods['planned_value'] < 0) | 
    (periods['earned_value'] < 0) | 
    (periods['actual_cost'] < 0)
]
print(f"Rows with negative values: {len(negatives)}")
if len(negatives) > 0:
    print(negatives[['period_number', 'planned_value', 'earned_value', 'actual_cost']])

# ─── 6. Plot cost curves for C2011-13 ────────────────────────────────────────
print("\n" + "=" * 60)
print("STEP 6: Visualize cost curves")
print("=" * 60)

plt.figure(figsize=(12, 6))
plt.plot(periods['period_number'], periods['cumulative_pv'], label='Planned Value (PV)', linewidth=2)
plt.plot(periods['period_number'], periods['cumulative_ev'], label='Earned Value (EV)', linewidth=2)
plt.plot(periods['period_number'], periods['cumulative_ac'], label='Actual Cost (AC)', linewidth=2)
plt.xlabel('Period')
plt.ylabel('Cumulative Cost (€)')
plt.title('C2011-13 Wind Farm: Cost Curves')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('outputs/figures/cost_curves_C2011-13.png')
plt.show()
print("Saved plot to outputs/figures/cost_curves_C2011-13.png")

# ─── 7. Calculate CPI and SPI trends ─────────────────────────────────────────
print("\n" + "=" * 60)
print("STEP 7: CPI and SPI trends")
print("=" * 60)

# Calculate CPI and SPI
periods['cpi'] = periods['earned_value'] / periods['actual_cost'].replace(0, pd.NA)
periods['spi'] = periods['earned_value'] / periods['planned_value'].replace(0, pd.NA)

print(f"CPI at period 1: {periods['cpi'].iloc[0]:.3f}")
print(f"CPI at period 119: {periods['cpi'].iloc[-1]:.3f}")
print(f"SPI at period 1: {periods['spi'].iloc[0]:.3f}")
print(f"SPI at period 119: {periods['spi'].iloc[-1]:.3f}")

# ─── 8. Plot CPI/SPI over time ───────────────────────────────────────────────
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(periods['period_number'], periods['cpi'], 'b-', linewidth=1.5)
plt.axhline(y=1, color='r', linestyle='--', alpha=0.5)
plt.xlabel('Period')
plt.ylabel('CPI')
plt.title('Cost Performance Index (CPI)')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(periods['period_number'], periods['spi'], 'g-', linewidth=1.5)
plt.axhline(y=1, color='r', linestyle='--', alpha=0.5)
plt.xlabel('Period')
plt.ylabel('SPI')
plt.title('Schedule Performance Index (SPI)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('outputs/figures/cpi_spi_trends_C2011-13.png')
plt.show()
print("Saved plot to outputs/figures/cpi_spi_trends_C2011-13.png")

# ─── 9. Summary statistics ───────────────────────────────────────────────────
print("\n" + "=" * 60)
print("STEP 8: Summary Statistics")
print("=" * 60)

print(f"Project: C2011-13 Wind Farm")
print(f"Budget at Completion: €{projects[projects['project_name'].str.contains('Wind', na=False)]['budget_at_completion'].values[0]:,.0f}")
print(f"Total Periods: {len(periods)}")
print(f"Final Cumulative PV: €{periods['cumulative_pv'].iloc[-1]:,.0f}")
print(f"Final Cumulative EV: €{periods['cumulative_ev'].iloc[-1]:,.0f}")
print(f"Final Cumulative AC: €{periods['cumulative_ac'].iloc[-1]:,.0f}")
print(f"Final Cost Variance: €{periods['cumulative_ac'].iloc[-1] - periods['cumulative_pv'].iloc[-1]:,.0f}")
```
